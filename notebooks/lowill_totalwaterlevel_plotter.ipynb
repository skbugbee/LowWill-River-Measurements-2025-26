## Total Water Level Plotter
This script accompanies the `lowill_totalwaterlevel_dataset.csv` found in the https://github.com/skbugbee/LowWill-Measurements-2025-26.git GitHub repository.

To run this script, download the `lowill_totalwaterlevel_dataset.csv` to your computer and paste your filepath in the cell below. An example will be used as a placeholder.

```python
import pandas as pd
import matplotlib.pyplot as plt

twl_file = "/Users/sashabugbee/Downloads/lowill_totalwaterlevel_dataset.csv"
twl_df = pd.read_csv(twl_file, skiprows=20, encoding='latin-1', low_memory=False)

datetime_cols = [col for col in twl_df.columns if 'DATETIME' in col]
for col in datetime_cols:
    twl_df[col] = pd.to_datetime(twl_df[col], format='%m/%d/%y %H:%M', errors='coerce')
    
twl_df.head()
```

#### **Willapa River Timestack**

You can change the `plot_start` and `plot_end` variables to zoom in on different time periods. The December 2025 storm period is shaded grey for emphasis.

```python
fig, ax = plt.subplots(7, 1, figsize=(8, 12), sharex=True)
plot_start = pd.to_datetime('2025-12-01')
plot_end = pd.to_datetime('2025-12-24')

sensors = [
    ('WR08', 'forestgreen'),
    ('WR07', 'darkcyan'),
    ('WR06', 'steelblue'),
    ('WR04', 'royalblue'),
    ('WR03', 'mediumblue'),
    ('WR02', 'slateblue'),
    ('WR01', 'indigo'),
]

for i, (sensor, color) in enumerate(sensors):
    dt_col = f'{sensor} DATETIME [PST]'
    twl_col = [c for c in twl_df.columns if c.startswith(sensor) and 'TWL' in c][0]
    dt = twl_df[dt_col]
    twl = twl_df[twl_col]
    mask = (dt >= plot_start) & (dt <= plot_end)
    ax[i].plot(dt[mask], twl[mask], color=color, label=sensor)
    ax[i].legend(loc='upper right')
    ax[i].grid()
    ax[i].set_ylabel('TWL [m] (relative)' if sensor == 'WR08' else 'TWL NAVD88 [m]', fontsize=8)
    if sensor != 'WR08':
        ax[i].set_ylim([-1.5, 6.7])

shade_start = pd.Timestamp('2025-12-08')
shade_end = pd.Timestamp('2025-12-13')
for a in ax:
    a.set_xlim([plot_start, plot_end])
    a.axvspan(shade_start, shade_end, alpha=0.3, color='lightgrey', zorder=0)
ax[0].set_title('Willapa River TWL Timestack')

fig.tight_layout()
```
